### Exploratory Data Analysis (EDA)
There was a lot of EDA performed in the data gathering & cleaning portion of this project. This EDA will seek to further explore the relationships between features, and perform a final preprocessing of the data to prepare for modeling.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/quarterbacks_final.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# rearrange columns
cols = ['Name', 'Year', 'Month', 'Week', 'Team', 'Oppt', 'Age', 'h/a', 'Favored', 'Spread', 'O/U', 'Temperature', 
        'Wind', 'Weather_DOME', 'Weather_Fog', 'Weather_Rain', 'Weather_Rain | Fog', 'Weather_Snow',
        'Weather_Snow | Fog', 'Weather_Snow | Freezing Rain', 'Weather_Sunny', 'Attempts', 'Completions', 
        'Interceptions', 'Rating', 'TDs', 'Y/A', 'Yards', 'Opp_Avg_Att_Allowed', 'Opp_Avg_Comp_Allowed', 
        'Opp_Avg_Ints', 'Opp_Rank', 'Opp_Avg_TDs_Allowed', 'Opp_Avg_Yds_Allowed', 'FD salary', 'FD points']

df = df[cols]
```

```python
# group by player, year, week, and matchup
df = df.groupby(['Name', 'Year', 'Week', 'Month', 'Team', 'Oppt']).sum()
```

```python
df.shape
```

```python
sns.heatmap(df.corr())
```

> Liiiittle bit of multicollinearity in the data is what I'm gathering from this.

```python
sns.distplot(df['FD points']);
```

```python
df['FD points'].describe()[1:]
```

> Our target is pretty normally distributed.

```python
sns.distplot(df['FD salary']);
```

```python
df['FD salary'].describe()[1:]
```

> Let's break down the heat map above and look at each "category" of features as they relate to `FD points`

```python
# betting info versus FD points
betting_cols = ['Age', 'h/a', 'Favored', 'Spread', 'O/U', 'FD points']
sns.heatmap(df[betting_cols].corr(), annot = True)
```

```python
# player stats relationship with FD points
player_stat_columns = ['Attempts', 'Completions', 'Interceptions', 'Rating', 'TDs', 'Y/A', 'Yards', 'FD points']
sns.heatmap(df[player_stat_columns].corr(), annot = True)
```

```python
# player stats relationship with FD points
weather_cols = ['Temperature', 'Wind', 'Weather_DOME', 'Weather_Fog', 'Weather_Rain', 'Weather_Rain | Fog',
                'Weather_Snow', 'Weather_Snow | Fog', 'Weather_Snow | Freezing Rain', 'Weather_Sunny', 'FD points']

sns.heatmap(df[weather_cols].corr())
```

```python
def_stat_cols = ['Opp_Avg_Att_Allowed', 'Opp_Avg_Comp_Allowed', 'Opp_Avg_Ints', 
                 'Opp_Rank', 'Opp_Avg_TDs_Allowed', 'Opp_Avg_Yds_Allowed', 'FD points']
sns.heatmap(df[def_stat_cols].corr(), annot = True)
```

```python
sns.heatmap(df[['FD points', 'FD salary']].corr(), annot = True)
```

> So `FD points` is not highly correlated with ANY columns. This tells me that a linear regression is not going to do well. The predictors should have a linear relationship with the target variable.

```python
sns.regplot(df['FD points'], df['FD salary'], scatter_kws = {'alpha': 0.2}).set_title('Points scored based on Salary');
```

```python
sns.regplot(df['FD points'], df['Yards'], scatter_kws = {'alpha': 0.2}).set_title('Points scored based on Yards');
```
